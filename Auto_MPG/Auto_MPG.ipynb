   ### Step 3. Assign each to a variable called cars1 and cars2

```python
path1 = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv'
path2 = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv'
cars1 = pd.read_csv(path1)
cars2 = pd.read_csv(path2)
```

```python
cars = pd.concat([cars1,cars2],ignore_index=True)
cars
```
