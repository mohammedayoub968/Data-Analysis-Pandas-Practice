### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic

```python
path = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
titanic = pd.read_csv(path)
titanic.head()
```

### Step 4. Set PassengerId as the index

```python
titanic.set_index('PassengerId' , inplace=True)
```

```python
titanic.head()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
sex_counts = titanic['Sex'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightpink'], startangle=90)
plt.title('Male / Female Proportion')
plt.axis('equal')
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=titanic, x='Age', y='Fare', hue='Sex', palette={'male': 'blue', 'female': 'pink'})

plt.title('Fare vs Age by Gender')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

### Step 7. How many people survived?

```python
titanic['Survived'].sum()
```

### Step 8. Create a histogram with the Fare payed

```python
plt.figure(figsize=(8, 5))
plt.hist(titanic['Fare'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Fare Paid')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```
